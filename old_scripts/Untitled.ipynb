```python
def loop_plot(x,plots):
    figs={}
    axs={}
    for idx,plot in enumerate(list(plots.columns)):
        y=plots.loc[:,plot]
        figs[idx]=plt.figure()
        axs[idx]=figs[idx].add_subplot(111)
        axs[idx].bar(x, y, width=1)
        axs[idx].set_ylabel = plot
        if idx>0:
            axs[idx].spines['top'].set_visible(False)
    return figs, axs
```

```python
def differentHaploBlock(block1, block2, maxDifferentSNPs):
    if (len(block1)!=len(block2)):
        return
    else:
        matches = 0
        for i, j in zip(block1, block2):
            if i == j:
                matches+=1
        if len(block1)-matches<maxDifferentSNPs:
            return True
```

```python
def ifpresent(new_item, list, maxDifferentSNPs):
    presence = 0
    for item in list:
        if differentHaploBlock(new_item, item, maxDifferentSNPs):
            presence = 0
        else:
            presence = 1
    return(presence)
```
